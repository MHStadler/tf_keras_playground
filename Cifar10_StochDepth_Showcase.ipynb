```python
import pandas as pd

constdepth_hist = pd.read_csv('./trained_models/cifar10_resnet32_constdepth_hist.csv')
resdrop_hist = pd.read_csv('./trained_models/cifar10_resnet32_resdrop_hist.csv')
wrapper_hist = pd.read_csv('./trained_models/cifar10_resnet32_wrapper_hist.csv')
```

```python
import matplotlib.pyplot as plt

plt.figure()

plt.plot(constdepth_hist['val_loss'][80:], label = 'Const. Depth')
plt.plot(resdrop_hist['val_loss'][80:], label = 'Stoch. Depth (Layer)')
plt.plot(wrapper_hist['val_loss'][80:], label = 'Stoch. Depth (Wrapper)')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Val. Loss')
plt.show()
```
